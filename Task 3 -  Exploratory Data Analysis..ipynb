```python
print("there are" ,df.State.nunique(),"States")
```

```python
x=df['State']
y=df['Sales']
fig,ax = plt.subplots(figsize =(40,10))
ax.bar(x,y)
fig.autofmt_xdate()
ax.set_xlabel('State')
ax.set_ylabel('Sales')
ax.set_title('State wise Sales distribution')
plt.show()
```

# State wise Sales Distribution

```python
x=df['Region']
y=df['Sales']
fig,ax = plt.subplots(figsize =(40,10))
ax.bar(x,y)
fig.autofmt_xdate()
ax.set_xlabel('Region', size = 8)
ax.set_ylabel('Sales', size = 8)
ax.set_title('Region wise Sales distribution')
plt.show()
```
